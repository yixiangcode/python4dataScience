```python
# student_name = "Cheng Yi Xiang"
# student_id  = "B230031A"
# title = "BTPR3203 Python for Data Science Assignment 1"

import datetime

beneficiary_id = 1

inventory_stock = {
    'Food Package': 100,
    'Hygiene Kit': 50,
    'Blanket': 30
}

beneficiary_list = []

distribution_history = []

def update_stock(item_name, quantity):
  if item_name in inventory_stock.keys():
    print(f"Stock for {item_name} updated. New quantity: {quantity}.")
  else:
    print(f"New item {item_name} added with quantity: {quantity}.")
  inventory_stock[item_name] = quantity
  input("Press any key to continue...")

def display_inventory():
  print("------------------ Current Inventory -------------------")
  for i, j in inventory_stock.items():
    print(f"{i}: {j} units")
  input("Press any key to continue...")

def register_beneficiary(beneficiary_name):
  global beneficiary_id
  beneficiary = {
      "beneficiary_name" : beneficiary_name,
      "beneficiary_id" : beneficiary_id,
      "eligibility_status" : True
  }
  beneficiary_list.append(beneficiary)
  print(f"Beneficiary registered. ID: {beneficiary_id}, Name: {beneficiary_name}")
  beneficiary_id += 1
  input("Press any key to continue...")

def update_eligibility(beneficiary_id, eligibility_status):
  for beneficiary in beneficiary_list:
    if beneficiary["beneficiary_id"] == beneficiary_id:
      beneficiary["eligibility_status"] = eligibility_status
      print(f"Beneficiary eligibility updated. ID: {beneficiary_id}, Status: {eligibility_status}")
      input("Press any key to continue...")
      return
  print(f"Beneficiary ID {beneficiary_id} not found.")
  input("Press any key to continue...")

def distribute_item(beneficiary_id, item_name, quantity_to_distribute):
  for beneficiary in beneficiary_list:
    if beneficiary["beneficiary_id"] == beneficiary_id:
      if beneficiary["eligibility_status"] == True:
        if item_name not in inventory_stock.keys():
          print(f"Item {item_name} not found in inventory.")
          input("Press any key to continue...")
          return
        if inventory_stock[item_name] < quantity_to_distribute:
          print(f"Insufficient stock for {item_name}. Available: {inventory_stock[item_name]}")
          input("Press any key to continue...")
          return
        inventory_stock[item_name] -= quantity_to_distribute
        distribution_history.append((datetime.datetime.now().strftime('%Y-%m-%d %H:%M:%S'), beneficiary_id, item_name, quantity_to_distribute));
        print(f"Successfully distributed {quantity_to_distribute} of {item_name} to Beneficiary ID {beneficiary_id}.")
        input("Press any key to continue...")
      else:
        print(f"Beneficiary ID {beneficiary_id} is not eligible for distribution.")
        input("Press any key to continue...")
      return
  print(f"Beneficiary ID {beneficiary_id} not found.")
  input("Press any key to continue...")

def generate_reports():
  print("\n------------------- Report Summary ---------------------")
  total_beneficiaries = []
  for i in distribution_history:
    if i[1] not in total_beneficiaries:
      total_beneficiaries.append(i[1])
  print("Total unique beneficiaries served:", len(total_beneficiaries))

  total_items = {}
  for i in distribution_history:
    item = i[2]
    quantity = i[3]
    if item in total_items:
      total_items[item] += quantity
    else:
      total_items[item] = quantity
    print(f"Time: {i[0]}, Beneficiary ID: {i[1]}, Item: {i[2]}, Quantity: {i[3]}")

  print("Total quantity distributed by item:")
  for item in total_items:
      print(f"{item}: {total_items[item]} units")

  out_of_stock_items = []
  for item in inventory_stock:
    if inventory_stock[item] == 0:
      out_of_stock_items.append(item)
  if len(out_of_stock_items) == 0:
    print("All items are currently in stock.")
  else:
    print("Items out of stock:")
    for item in out_of_stock_items:
      print(item)
  input("Press any key to continue...")

while True:
  print("--------------------------------------------------------")
  print("1. Add/Update Item Stock\n2. Display Current Inventory\n3. Beneficiary Management (Part 2)\n4. Distribution (Part 3)\n5. Reports (Part 4)\n0. Exit Program")
  option = int(input("Please enter your option: "))

  if option == 1:
    item_name = input("Please enter the item's name: ")
    quantity = int(input("Please enter the item's quantity: "))
    update_stock(item_name.title(), quantity)

  elif option == 2:
    display_inventory()

  elif option == 3:
    print("\n------------------- Beneficiary Menu ---------------------\n1. Register New Beneficiary\n2. Update Beneficiary Eligibility")
    beneficiary_option = int(input("Please enter your option: "))
    if beneficiary_option == 1:
      beneficiary_name = input("Please enter the beneficiary's name: ")
      register_beneficiary(beneficiary_name)
    elif beneficiary_option == 2:
      id = int(input("Please enter the beneficiary ID: "))
      eligibility_status = input("Please enter the eligibility status (True/False): ")
      update_eligibility(id, eligibility_status.title())

  elif option == 4:
    id = int(input("Please enter the beneficiary ID: "))
    item_name = input("Please enter the item name: ")
    quantity_to_distribute = int(input("Please enter the quantity to distribute: "))
    distribute_item(id, item_name.title(), quantity_to_distribute)

  elif option == 5:
    generate_reports()

  else:
    break
```
